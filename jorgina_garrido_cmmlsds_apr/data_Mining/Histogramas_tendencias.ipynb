```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ENS_2017_valida_2 = pd.read_csv("../data/Bases_trabajo/ENS_2017_valida_2.csv", sep =',')
```

```python

for i in ENS_2017_valida_2.columns:
    plt.hist(x=ENS_2017_valida_2[i], bins=5, color='#F2AB6D', rwidth=0.85)
    plt.title(i)
    plt.xlabel(i)
    plt.ylabel('Frecuencia')

    plt.savefig("../reports/Histogramas/i.jpg", bbox_inches='tight')

    plt.show()


    
```
