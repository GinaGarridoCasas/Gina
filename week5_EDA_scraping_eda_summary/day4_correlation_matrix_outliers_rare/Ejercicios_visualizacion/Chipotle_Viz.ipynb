### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

### Step 4. Create a histogram of the top 5 items bought

```python
plt.figure(figsize=(6,5))
chipo_1 = chipo['item_name'].value_counts()[:5].sort_values(ascending=True)
order = chipo_1.index
sns.countplot(data=chipo, x='item_name', color = 'skyblue', order=order)
plt.title("Most ordered Chipotle's Items")
plt.ylabel('Number of Times Ordered')
plt.xlabel('items')
plt.xticks(rotation = 90)


```
