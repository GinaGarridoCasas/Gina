### Step 1. Import the necessary libraries

```python

# remember to %matplotlib inline
import pandas as pd
import matplotlib as mpl # esto nunca usa
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
datos = {'first_name' : ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 'last_name':['MIller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 'age': [42,52,36, 24,73], 'female':[0,1,1,0,1], 'preTestScore': [4,24,31,2,3], 'postTestScore': [25,94,57,62,70]}

data = pd.DataFrame(datos)
data
```

```python
data.info()
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(data['preTestScore'],data['postTestScore'], data['age'], alpha = 0.9)
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")
plt.title ("PreTestScore x PostTestScore")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.relplot(data=data,
            x="preTestScore",
            y="postTestScore",
            hue="female", size = 'female',
            sizes = (450, 100)
            alpha = 0.5, palette= 'muted',height=6, legend = False)

plt.xlabel("preTestScore")
plt.ylabel("postTestScore")
plt.title ("PreTestScore x PostTestScore")


```
