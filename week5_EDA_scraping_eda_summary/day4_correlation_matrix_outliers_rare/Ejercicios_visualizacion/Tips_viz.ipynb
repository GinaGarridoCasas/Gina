### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('data/tips.csv')
tips
```

```python
tips.info()
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop('Unnamed: 0', inplace = True, axis=1)

tips

```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(10,5))
sns.distplot(tips['total_bill'], hist = True)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title ("Total Bill")
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips['total_bill'],tips['tip'], sizes = (45,), alpha = 0.9)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title ("Total Bill and Tip relationship")
```

### Step 7.  Create one image with the relationship of `total_bill`, `tip` and `size`.
#### Hint: It is just one function.

```python
plt.figure(figsize=(12,8))

sns.scatterplot(tips['total_bill'],tips['tip'], hue = tips['size'], sizes = (90,), alpha = 0.9)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title ("Total Bill and Tip relationship")

plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.scatter(x = tips['day'],y = tips['total_bill'])
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title ("Days and Total bill relationship")


plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = tips["tip"], y = tips["day"], hue = tips['sex'], s = 150)
plt.xlabel("Tip")
plt.ylabel("Day")
plt.title ("Tip and Day relationship by sex")


plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.catplot(x="day", y="total_bill", hue="time", kind="box", data=tips)
plt.xlabel("Day")
plt.ylabel("Total bill")
plt.title ("Day and Total bill relationship by sex. Boxplot")
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(10,5))
tips['tip'] = tips.tip.apply(lambda x: int(x))
sns.displot(data = tips, x ='tip', col ="time", kind = "hist", bins = 10)

```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
sns.relplot(
    data=tips, x="total_bill", y="tip",
    col="sex", hue="smoker", style="smoker",
    kind="scatter"
)
```
